```python
import numpy as np
import pandas as pd
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame from the Iris dataset
iris_data = pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['target'])
iris_data['target'] = iris_data['target'].astype(int)

# Visualize the classes using Matplotlib
plt.figure(figsize=(8, 6))
plt.scatter(iris_data['petal length (cm)'], iris_data['petal width (cm)'], c=iris_data['target'], cmap='viridis', edgecolor='k', s=40)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Iris Dataset - Petal Length vs Petal Width')
plt.show()

# Visualize the classes using Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='target', data=iris_data, palette='viridis', edgecolor='k', s=80)
plt.title('Iris Dataset - Petal Length vs Petal Width')
plt.show()
```
